```python
!pip install spacy

# Cargar el modelo de spaCy en español
!python -m spacy download es_core_news_sm
nlp = spacy.load("es_core_news_sm")
import spacy
# Leer el discurso desde el archivo
with open("discurso.txt", "r", encoding="utf-8") as file:
    discurso = file.read()

# Procesar el texto con spaCy
doc = nlp(discurso)

for l in doc.sents:
    print(l)

for l in doc.ents:
    print(l.text, l.label)
# Iterar sobre cada oración en el documento
for i, sent in enumerate(doc.sents, 1):
    # Imprimir la oración y el número de elementos
    print(f"Oración {i}: {sent.text}")
    print(f"Número de elementos: {len(sent.text)}")
    print()
```
